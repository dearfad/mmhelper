```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('txtdata.csv')
```

```python
data['reason']
```

```python
pattern = '\d{4}年\d{1,2}月\d{1,2}日'

for i in range(1102):
    txt = str(data.iloc[i]['reason'])
    result = re.search(pattern,txt)
    if result:
        text = result.group()
        data.loc[i,'date'] = text
    else:
        print(i, txt)

```
