```python
import pandas as pd
from bs4 import BeautifulSoup

data = pd.read_csv('pdpage.csv')
```

```python
data
```

```python
file = []
```

```python
for i in range(1102):
    soup = BeautifulSoup(data.iloc[i].page, 'lxml')    
    
    if soup.find_all('a'):        
        itemlist = soup.find_all('a')
        link  = []
        for item in itemlist:
            link.append(item['href'])
        file.append(link)
    elif soup.find_all('img'):
        itemlist = soup.find_all('img')
        link = []
        for item in itemlist:
            link.append(item['src'])
        file.append(link)
    else:
        file.append(None)
```

```python
len(file)
```

```python
data['file'] = file
```

```python
data
```

```python
data.to_csv('pdpage_file.csv', encoding='utf_8_sig', index=0)
```
