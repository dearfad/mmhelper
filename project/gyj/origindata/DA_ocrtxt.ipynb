```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('tempdata.csv')
```

```python
jpglist = []

pattern = '依据.*?如下处罚'

for i in range(1102):
    link = data.iloc[i]['link']
    type = data.iloc[i]['type']
    id = data.iloc[i]['id']
    reasontxt = data.iloc[i]['reasontxt']
    resulttxt = data.iloc[i]['resulttxt']
    basis = data.iloc[i]['basis']
    ocrtxt = data.iloc[i]['ocrtxt']
    
    if type=='jpg':
        jpglist.append(i)
        
        if ocrtxt is not np.nan:
            result = re.search(pattern, ocrtxt)
            
            if result:
                basis = result.group()
                print(id, basis)
                jpglist.remove(i)
                data.loc[i,'basis'] = basis
                data.loc[i,'reasontxt'] = ocrtxt.split(basis)[0]
                print(ocrtxt.split(basis)[0])
                data.loc[i,'resulttxt'] = ocrtxt.split(basis)[1]
                print(ocrtxt.split(basis)[1])
                
```
