```python
pattern = '足纪字.*?号'

for i in range(1102):
    
    id = data.iloc[i]['id']
    
    type = data.iloc[i]['type']
    
    pdf = data.iloc[i]['pdf']
    
    if type=='pdf':
        result = re.search(pattern, pdf)
        if result:
            text = result.group(0)
#             if '（' in text: text = text.replace('（','[')
#             if '）' in text: text = text.replace('）',']')
#             if '〔' in text: text = text.replace('〔','[')
#             if '〕' in text: text = text.replace('〕',']')
#             if '﹝' in text: text = text.replace('﹝','[')
#             if '﹞' in text: text = text.replace('﹞',']')
            
#             if '(' in text: text = text.replace('(','[')
#             if ')' in text: text = text.replace(')',']')
                 
                
            newdata.loc[i, 'sn'] = text
#             print(id, text, newdata.loc[i, 'sn'])
```
