```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('tempdata.csv')
olddata = pd.read_csv('pddata_titlefix.csv')
```

```python
jpglist = []

pattern = '依据.*?之规定'

for i in range(1102):
    link = data.iloc[i]['link']
    type = data.iloc[i]['type']
    id = data.iloc[i]['id']
    reasontxt = data.iloc[i]['reasontxt']
    resulttxt = data.iloc[i]['resulttxt']
    basis = data.iloc[i]['basis']
    ocrtxt = data.iloc[i]['ocrtxt']
    pdf = olddata.iloc[i]['pdf']
    
    if type=='pdf' and i!=819:
        print(id,'='*50)
        pdftxt = eval(pdf)
        pdftxt = pdftxt.replace(' \n\n','')
        pdftxt = pdftxt.replace(' ','')
        pdftxt = pdftxt.replace('\x0c','')
        pdftxt = pdftxt.replace('\n','')
        jpglist.append(i)
        
        result = re.search(pattern, pdftxt)
            
        if result:
            basis = result.group()
            print(id, basis)
            jpglist.remove(i)
            data.loc[i,'basis'] = basis
            data.loc[i,'reasontxt'] = pdftxt.split(basis)[0]
            print(pdftxt.split(basis)[0])
            data.loc[i,'resulttxt'] = pdftxt.split(basis)[1]
            print(pdftxt.split(basis)[1])
                
```

```python
for i in jpglist:
    pdftxt = olddata.iloc[i]['pdf']
    print(i, pdftxt)
```

```python
itemlist = jpglist

pattern = '根据.*?以下处罚'

for i in itemlist:
    link = data.iloc[i]['link']
    type = data.iloc[i]['type']
    id = data.iloc[i]['id']
    reasontxt = data.iloc[i]['reasontxt']
    resulttxt = data.iloc[i]['resulttxt']
    basis = data.iloc[i]['basis']
    ocrtxt = data.iloc[i]['ocrtxt']
    
    if type=='jpg':
        
        if ocrtxt is not np.nan:
            result = re.search(pattern, ocrtxt)
            
            if result:
                basis = result.group()
                print(id, basis)
                jpglist.remove(i)
                data.loc[i,'basis'] = basis
                data.loc[i,'reasontxt'] = ocrtxt.split(basis)[0]
                print(ocrtxt.split(basis)[0])
                data.loc[i,'resulttxt'] = ocrtxt.split(basis)[1]
                print(ocrtxt.split(basis)[1])
    
```
