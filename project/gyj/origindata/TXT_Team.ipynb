```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('txtdata.csv')
```

```python
pattern = '场.*?队与.*?队'

for i in range(1102):
    print(i)
    txt = str(data.iloc[i]['reason'])
    result = re.search(pattern,txt)
    if result:
        text = result.group()
        if ')' in text:
            teams = text.split(',')[-1]
        elif '，' in text:
            teams = text.split('，')[1]
        elif ',' in text:
            teams = text.split(',')[1]
        else:
            teams = text.strip('场')
        
        try:
            team1, team2 = teams.split('与')
            print(team1, team2)
            data.loc[i,'team1'] = team1
            data.loc[i,'team2'] = team2
        except:
            pass

```
