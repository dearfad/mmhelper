```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('txtdata.csv')
```

```python
pattern = '上述.*?规定执行'

prule = '《.*?》'

pitem = '第.*?条'

for i in range(1102):

    text = data.iloc[i]['result']
    if text is not np.nan:
        result = re.search(pattern,text)
        if result:
            txt = result.group()
            pdcontent = text.split(txt)[0]
            print(i, pdcontent)
            data.loc[i,'pdcontent'] = pdcontent
            pdrule = re.findall(prule,txt)
            if pdrule:
                print(i, pdrule)
            data.loc[i,'pdrule'] = str(pdrule)

            pditem = re.findall(pitem,txt)
            if pdrule:
                print(i, pditem)
            data.loc[i,'pditem'] = str(pditem)
    

            
        else:
            print(i)
```
