```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('txtdata.csv')
```

```python
pattern = '\w{4}年\w{1,2}月\w{1,2}日'

for i in range(1102):

    text = data.iloc[i]['result']
    if text is not np.nan:
        result = re.findall(pattern,text)
        if result:
            print(i, result[-1])
            data.loc[i,'pddate'] = result[-1]
        else:
            print(i)
```
