```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('txtdata.csv')
```

```python
pattern = '在.*?行'

for i in range(1102):
    reason = data.iloc[i]['reason']
    if reason is not np.nan:
        result = re.search(pattern, reason)
        if result:
            text = result.group()
            text = text.strip('在')
            text = text.strip('举行')
            text = text.strip('进行')
            print(i,text)
            data.loc[i,'field'] = text
```
