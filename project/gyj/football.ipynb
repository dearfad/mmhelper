# 中国足球协会处罚分析

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')
```

## 一、受罚行为分类统计分析

### 1.动作侵犯行为数量统计（球员、官员）

#### 1.1 肢体接触动作侵犯行为数量统计

```python
# t1 = 1
touchdata = data[data['t1']==1]
```

```python
touchdata.shape
```

* 1.1.1 各年 球员数量
* 1.1.2 各年 官员数量

```python
touchdata.groupby('year').size()
```

```python
# 各年总数量
sns.countplot(x='year', data=touchdata);
```

```python
touchdata.groupby(['year','受罚者']).size()
```

```python
# 各年分类数量
sns.countplot(x='year',hue='受罚者',data=touchdata);
```

* 1.1.3 各年 球员侵犯对方球员数量
* 1.1.4 各年 球员侵犯裁判员数量
* 1.1.5 各年 球员侵犯对方官员数量

```python
tempdata = touchdata[touchdata['受罚者']=='运动员']
temp1 = tempdata[tempdata['运动员']==1].copy()
temp1['type'] = '运动员'
temp2 = tempdata[tempdata['裁判员']==1].copy()
temp2['type'] = '裁判员'
temp3 = tempdata[tempdata['官员']==1].copy()
temp3['type'] = '官员'
tempdata = pd.concat([temp1,temp2,temp3])
sns.countplot(x='year',hue='type',data=tempdata);
```

* 1.1.3 各年 官员侵犯对方球员数量
* 1.1.4 各年 官员侵犯裁判员数量
* 1.1.5 各年 官员侵犯对方官员数量

```python
tempdata = touchdata[touchdata['受罚者']=='官员']
temp1 = tempdata[tempdata['运动员']==1].copy()
temp1['type'] = '运动员'
temp2 = tempdata[tempdata['裁判员']==1].copy()
temp2['type'] = '裁判员'
temp3 = tempdata[tempdata['官员']==1].copy()
temp3['type'] = '官员'
tempdata = pd.concat([temp1,temp2,temp3])
sns.countplot(x='year',hue='type',data=tempdata);
```
