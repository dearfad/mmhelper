```python
import pandas as pd

data = pd.read_csv('pdpage_file.csv')

datafile = data.dropna()
```

```python
datafile
```

```python
import requests
import time

header = {
    'Connection': 'keep-alive',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36',
    'Accept': 'text/css,*/*;q=0.1',
    'Accept-Encoding': 'gzip, deflate',
    'Accept-Language': 'zh-CN,zh;q=0.9,en;q=0.8,ja;q=0.7,zh-TW;q=0.6'
}

timeout = 5
```

```python

for i in range(681):

    print(i)
    
    filelinks = datafile.iloc[i].file
    
    filelinks = filelinks.strip("['")
    
    filelinks = filelinks.strip("']")
    
    filelinks = filelinks.split("', '")
    
    id = datafile.iloc[i].id

    url = filelinks[0]
    if 'http:' not in url:
        url = 'http://www.thecfa.cn' + url
    suffix = url.split('.')[-1]
    name = str(id) + '.' + suffix
    
    result = False
    
    while not result:

        try:
            print(url, name)
            page = requests.get(url, headers = header, timeout = timeout)
            with open(name, 'wb') as f:
                f.write(page.content)
                print(id, 'OK')
                result = True
        except:
            time.sleep(2)
    
    if len(filelinks)==2:
        url = filelinks[1]
        if 'http:' not in url:
            url = 'http://www.thecfa.cn' + url
        suffix = url.split('.')[-1]
        name = str(id) + '_2' + '.' + suffix
        
        result = False
    
        while not result:

            try:
                print(url, name)
                page = requests.get(url, headers = header, timeout = timeout)
                with open(name, 'wb') as f:
                    f.write(page.content)
                    print(id, '2_OK')
                    result = True
            except:
                time.sleep(2)
```
