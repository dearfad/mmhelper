```python
import pandas as pd
import re
import numpy as np

data = pd.read_csv('pddata_titlefix.csv')

newdata = pd.read_csv('tempdata.csv')
```

```python
data
```

```python
newdata['reason'] = np.nan
newdata['basis'] = np.nan
newdata['result'] = np.nan
newdata['type'] = data['type']
newdata
```

```python
pattern = '依据.*?如下处罚'

leftlist = []

for i in range(1102):
    
    type = data.iloc[i]['type']
    
    id = data.iloc[i]['id']
    
    link = newdata.iloc[i]['link']
    
    page = data.iloc[i]['page']
    
    
    if type=='txt':
        
        text = re.search(pattern, page)        
        
        
        if text:
            basis = text.group()
            reason = page.split(basis)[0]
            result = page.split(basis)[1]

            newdata.loc[i,'reason'] = reason
            newdata.loc[i,'basis'] = basis
            newdata.loc[i,'result'] = result

        else:
            print(id, link)
            leftlist.append(id)
```
