```python
import pandas as pd
import re
import numpy as np

data = pd.read_csv('pddata_titlefix.csv')

newdata = pd.read_csv('tempdata.csv')
```

```python
data
```

```python
newdata['reason'] = np.nan
newdata['basis'] = np.nan
newdata['result'] = np.nan
newdata
```

```python
pattern = '依据.*?如下处罚'

for i in range(1102):
    
    type = data.iloc[i]['type']
    
    id = data.iloc[i]['id']
    
    page = data.iloc[i]['page']
    
    
    if type=='txt':
        
        text = re.search(pattern, page)
        
        if text:
            basis = text.group()
            reason = page.split(basis)[0]
            result = page.split(basis)[1]
            print("="*80)
            print(id)
            print(reason)
            print('-'* 80)
            print(basis)
            print('-'* 80)
            print(result)
            newdata.loc[i,'reason'] = reason
            newdata.loc[i,'basis'] = basis
            newdata.loc[i,'result'] = result
```

```python
pattern = '依据.*?如下处罚'

for i in range(1102):
    
    type = data.iloc[i]['type']
    
    id = data.iloc[i]['id']
    

    
    if type=='img':
        
        ocr = data.iloc[i]['ocr']
    

    
        ocr = eval(ocr)
        
    
        ocr = ocr['words_result']
        
        print(ocr)
        
        text = re.search(pattern, ocr)
        
        if text:
            basis = text.group()
            reason = ocr.split(basis)[0]
            result = ocr.split(basis)[1]
            print("="*80)
            print(id)
            print(reason)
            print('-'* 80)
            print(basis)
            print('-'* 80)
            print(result)
            newdata.loc[i,'reason'] = reason
            newdata.loc[i,'basis'] = basis
            newdata.loc[i,'result'] = result
```
