```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('txtdata.csv')
```

```python
pattern = '《.*?》'
for i in range(1102):
    print(i)
    basis = data.iloc[i]['basis']
    if basis is not np.nan:
        results = re.findall(pattern, basis)
        if results:
            data.loc[i,'baset'] = str(results)
            print(results)
```

```python
pattern = '第.*?条'
for i in range(1102):
    print(i)
    basis = data.iloc[i]['basis']
    if basis is not np.nan:
        results = re.findall(pattern, basis)
        if results:
            data.loc[i,'baitem'] = str(results)
            print(results)
```
