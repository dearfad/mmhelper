```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('txtdata.csv')
```

```python
prsdata = '根据.*?等'
prscontent = ''
prsresult = '造成.*?影响.*?'

unlist = []

for i in range(1102):
    unlist.append(i)
    
    text = data.iloc[i]['reason']
    
    tag = 0
    
    result = re.search(prsdata, text)
    if result:
        data.loc[i,'rsdata'] = result.group()
        text = text.strip(result.group())
        unlist.remove(i)
        tag = 1
    
    result = re.search(prsresult, text)
    if result:
        data.loc[i,'rsresult'] = result.group()
        text = text.strip(result.group())
        if tag==0: unlist.remove(i)
        
    data.loc[i,'rscontent'] = text  
    
```
