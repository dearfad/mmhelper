```python
import pandas as pd
import numpy as np
import os.path
from pdfminer.high_level import extract_text

data = pd.read_csv('pdpage_ocr.csv')
```

```python
data
```

```python
data['pdf'] = np.nan
data['pdf_2'] = np.nan
```

```python
data
```

```python
for i in range(1102):
    
#     print(i)
    
    fileprefix = str(data.iloc[i]['id'])
    
    filename = fileprefix + '.pdf'
    
    if os.path.exists(filename):
        with open(filename, 'rb') as f:
            pdf = f.read()
        print(i, filename)
        text = extract_text(filename)
        pdftext = repr(text)
        data.loc[i,'pdf'] = pdftext
    
    filename2 = fileprefix + '_2.pdf'
    
    if os.path.exists(filename2):
        with open(filename2, 'rb') as f:
            pdf = f.read()
        print(i, filename2)
        text = extract_text(filename2)
        pdftext = repr(text)
        data.loc[i,'pdf_2'] = pdftext
```

```python
data
```

```python
data.to_csv('pdpage_pdf.csv', encoding='utf_8_sig', index=0)
```
