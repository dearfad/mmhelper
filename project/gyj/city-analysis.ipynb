# 赛区分析

```python
import pandas as pd
pd.options.display.max_columns = None # 显示所有列
# 读取 .csv 文件
origin_data = pd.read_csv('football.csv')
citydata = origin_data[origin_data.球队=='赛区']
print(f'赛区处罚共计: {citydata.shape[0]} 项')
citydata
```

```python
# tempcity = [ item.strip('赛区') for item in citydata['受罚']]

city = citydata['受罚'].value_counts()

citydict = dict(city)

newcity = {}

for item in citydict.keys():
    cityname = item.strip('赛区')
    newcity[cityname] = citydict[item]
    
cities = list(newcity.items())
```

```python
cities
```

```python
from pyecharts.charts import Map
```

```python
map = Map()
```

```python
map.add("", cities)
```

```python
map.render_notebook()
```
