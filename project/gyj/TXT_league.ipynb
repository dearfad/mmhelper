```python
import pandas as pd
import numpy as np
import re

data = pd.read_csv('txtdata.csv')
```

```python
overlist = []
pattern = "中国足球协会.*?联赛.*?第.*?轮场序.*?场"
for i in range(1102):
    print(i)
    overlist.append(i)
    reason = data.iloc[i]['reason']
    if reason is not np.nan:
        result = re.search(pattern, reason)
        if result:

            text = result.group()
            league = re.search('中国足球协会.*?联赛',text).group()
            round = re.search('第.*?轮',text).group()
            round = round.strip('第')
            round = round.strip('轮')
            order = re.search('场序.*?场',text).group()
            order = order.strip('场序')
            order = order.strip('场')
            print(i,league,round,order)
            data.loc[i, 'league'] = league
            data.loc[i, 'round'] = round
            data.loc[i, 'order'] = order
            overlist.remove(i)
            
            
```
