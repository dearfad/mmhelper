```python
from auto_ml import Predictor
from sklearn.model_selection import train_test_split
import pandas as pd

dataframe = pd.read_csv('./data/duct/duct.csv')

df_train, df_test = dataframe[:425], dataframe[425:]

```

```python
column_descriptions = {
    'group': 'output',
    'menopause': 'categorical',
    'gestation': 'categorical',
    'pain': 'categorical',
    'ectasia': 'categorical',
    'multiple': 'categorical',
    'distance': 'categorical',
    'size': 'categorical'
}

ml_predictor = Predictor(type_of_estimator='regressor', column_descriptions=column_descriptions)

ml_predictor.train(df_train)

ml_predictor.score(df_test, df_test.group)
```
